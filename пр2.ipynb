### Задание 1.1: Синусоида и косинусоида
Используя библиотеки NumPy и Matplotlib, постройте на одном графике функции синуса и косинуса на интервале от 0 до $2\pi$ Оформите график, добавив легенду и подписи к осям (`.legend()` `.ylabel()` `.xlabel()`)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# интервал
x = np.linspace(0, 2 * np.pi, 1000)
# вычисление функций
sin = np.sin(x)
cos = np.cos(x)
#построение графика
plt.plot(x, sin, label='sin(x)', color='blue')
plt.plot(x, cos, label='cos(x)', color='red')
#легенда и подписи
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функций синуса и косинуса на интервале от 0 до  2π')
plt.legend()

# Отображение графика
plt.grid(True)
plt.show()
```

### Задание 1.7: Интерактивный график линий

1. Используя библиотеку numpy, сгенерируйте одномерный массив, представляющий месяцы года (от 1 до 12).
2. Используя библиотеку numpy, сгенерируйте три одномерных массива, представляющих среднюю температуру воздуха в трех городах (например, Москва, Париж и Нью-Йорк) в соответствующие месяцы. Пусть температура варьируется в диапазоне от -10 до 30 градусов Цельсия.
3. Используя библиотеку plotly express, создайте интерактивный график линий с помощью функции [`plotly.express.line`](https://plotly.com/python-api-reference/generated/plotly.express.line.html).
4. В качестве параметров функции `line` передайте массив месяцев в качестве оси `x`, массивы температур в качестве оси `y`, а также названия городов в качестве параметра `color`, чтобы разделить линии по группам .
5. Настройте внешний вид графика, добавляя подписи к осям, заголовок графика и легенду [`.update_layout()`](https://plotly.com/python/creating-and-updating-figures/#updating-figure-layouts).

```python
import numpy as np
import plotly.express as px
import pandas as pd
month = np.arange(1, 13, 1)
Moscow = np.array([-10, -9, -2, 6, 13, 17, 18, 16, 11, 5, -1, -6])
Beijing = np.array([-6, -1, 6, 15, 20, 25, 27, 26, 20, 14, 5, -2])
Rome = np.array([7, 8, 11, 13, 18, 22, 25, 26, 21, 17, 12, 8])

data = {
    "Месяц": np.tile(month, 3),  # Повторяем массив месяцев для каждого города
    "Температура, °C": np.concatenate([Moscow, Beijing, Rome]),  # Объединяем температуры
    "Город": ["Москва"] * 12 + ["Пекин"] * 12 + ["Рим"] * 12  # Названия городов
}
df = pd.DataFrame(data)
fig = px.line(df, x="Месяц", y="Температура, °C", color="Город", title='Средняя температура воздуха по месяцам в трех городах')
fig.show()
```

### Задание 1.8: Интерактивный график столбцов

**Цель:** Визуализировать интерактивный график столбцов, отображающий распределение населения по странам мира.

1. Используя библиотеку numpy, сгенерируйте одномерный массив, представляющий названия 10 стран мира (например, Россия, Китай, США и т.д.).
2. Используя библиотеку numpy, сгенерируйте одномерный массив, представляющий население каждой из выбранных стран в миллионах человек. Пусть население варьируется в диапазоне от 10 до 1500 миллионов человек.
3. Используя библиотеку plotly express, создайте интерактивный график столбцов с помощью функции [`plotly.express.bar`](https://plotly.com/python-api-reference/generated/plotly.express.bar.html?highlight=plotly%20express%20bar#plotly.express.bar).
4. В качестве параметров функции `bar` передайте массив названий стран в качестве оси `x`, массив населения в качестве оси `y`, а также массив населения в качестве параметра `color`, чтобы изменять цвет столбцов в зависимости от их значения.
5. Настройте внешний вид графика, добавляя подписи к осям, заголовок графика и цветовую шкалу [`.update_layout()`](https://plotly.com/python/creating-and-updating-figures/#updating-figure-layouts).

```python
import plotly.express as px
```

```python
countries = np.array(['Россия', 'Аргентина', 'Германия', 'Австралия', 'Италия', 'Турция', 'Франция', 'Испания', 'Бельгия', 'США'])
population = np.array([144.2, 46.23, 83.8, 26.01, 58.94, 84.98, 67.97, 47.78, 11.69, 333.3])

fig = px.bar(x=countries, y=population, color=population, labels={'x': 'Страна', 'y': 'Население, млн'}, title='Население 10 стран мира в миллионах человек', color_continuous_scale='turbo')
fig.update_layout(title_text="Население 10 стран мира в миллионах человек",
                  title_font_size=30)
fig.show()
```
